```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table1 = pd.read_table("output/error-zero.txt")
table1 = table1.values
table2 = pd.read_table("output/error-random.txt")
table2 = table2.values
table3 = pd.read_table("output/error-better.txt")
table3 = table3.values
```

```python
plt.semilogy(table1, color=(1, 0.3, 0), label='zero-init')
plt.semilogy(table2, color=(0.2, 0.8, 0.2), label='rand-init')
plt.semilogy(table3, color=(0, 0.3, 0.9), label='guess-init')
plt.legend(fontsize=12)
```
